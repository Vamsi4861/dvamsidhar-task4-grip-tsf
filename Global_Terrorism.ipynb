```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1')
data.head()
```

```python
data.columns.values
```

```python
data.rename(columns={'eventid':'Event_ID','iyear':'Year','imonth':'Month','iday':'Day','region_txt':'Region','country_txt':'Country','provstate':'State',
                    'city':'City','latitude':'latitude','longitude':'longitude','attacktype1_txt':'AttackType','nkill':'kill',
                     'nwound':'wound','targtype1_txt':'target1','weaptype1_txt':'Weapon','summary':'Summary','gname':'Group'},inplace=True)
```

```python
data = data[['Year','Month','Day','Country','Region','State','City','latitude','longitude','AttackType','kill','wound','target1','Weapon','Group','motive','Summary']]
```

```python
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['wound']=data['wound'].fillna(0)
data['kill']=data['kill'].fillna(0)
```

```python
data['Casualitites']=data['wound']+data['kill']
```

```python
data.info()
```

```python
data.describe()
```

```python
year=data['Year'].unique()
years_count=data['Year'].value_counts(dropna=False).sort_index()
plt.figure(figsize=(18,10))
sns.barplot(x=year,y=years_count,palette="tab10")
plt.xticks(rotation=50)
plt.xlabel('Attacking Year',fontsize=20)
plt.yticks(rotation=0)
plt.ylabel('Number of Attacks each year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
pd.crosstab(data.Year,data.Region).plot(kind='area',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region in each year',fontsize = 30)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel('Year',fontsize=20)
plt.show()
```

```python
attack=data.Region.value_counts()[:10]
attack
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='YlOrBr_r')
plt.title('Top Countries Affected')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation=50)
plt.show()
```

```python
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(20,10))
df.plot(kind = 'bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title('People Died due to the Attack',fontsize=30)
plt.ylabel('Number of killed people',fontsize=20)
plt.xlabel('Year',fontsize=20)
top_side = ax4.spines['top']
top_side.set_visible(False)
right_side = ax4.spines['right']
right_side.set_visible(False)
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(20,10),color='blue')
plt.xticks(rotation=50)
plt.xlabel('City',fontsize=30)
plt.ylabel('Number of attack',fontsize=20)
plt.xlabel('Top 10 most effected city',fontsize=20)
plt.show()
```

```python
data['AttackType'].value_counts().plot(kind = 'bar',figsize = (20,10),color = 'green')
plt.xticks(rotation = 50)
plt.ylabel('Number of attack',fontsize=20)
plt.xlabel('AttackType',fontsize=20)
plt.title('Name of AttackType',fontsize=30)
plt.show()
```

```python
data[['AttackType','kill']].groupby(['AttackType'],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['purple'])
plt.xticks(rotation=50)
plt.xlabel('AttackType',fontsize=20)
plt.ylabel('Number of people',fontsize=20)
plt.title('Number of killed',fontsize=30)
plt.show()
```

```python
data[['AttackType','wound']].groupby(['AttackType'],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['orange'])
plt.xticks(rotation=50)
plt.title('No. of wounded',fontsize=30)
plt.ylabel('No. of people',fontsize=20)
plt.xlabel('Attack Type',fontsize=20)
plt.show()
```

```python
data['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color = 'cyan',figsize=(20,10))
plt.title('Top 10 terrorist group attack on the basis of number of attacks',fontsize=30)
plt.xlabel('terrorist group name',fontsize=20)
plt.ylabel('Attack number',fontsize=20)
plt.show()
```

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='indigo',figsize=(20,10))
plt.title('Top 10 terrorist group attack on the basis of Number of People killed',fontsize=30)
plt.xlabel('Terrorist group name',fontsize=20)
plt.ylabel('No. of people killed',fontsize=20)
plt.show()
```
